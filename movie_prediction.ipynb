```python
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credites = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
credites.head()
```

```python
movies.shape
```

```python
movies=movies.merge(credites,on='title')
```

```python
movies.shape
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

movies.head()

```python
movies.head()
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
import ast
def convert(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
    
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast'].values
```

```python
import ast
def convert3(text):
    l=[]
    count=0
    for i in ast.literal_eval(text):
        if count<3:
            l.append(i['name'])
        count+=1    
    return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(text):
    l=[]
    count=0
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            l.append(i['name'])
            break
          
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)





```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].str.split()
```

```python
movies.head()
```

```python
def collapse(text):
    l1=[]
    for i in text:
        l1.append(i.replace(" ",""))
    return l1
        
```

```python
movies['cast']=movies['cast'].apply(collapse)
movies['crew']=movies['crew'].apply(collapse)
movies['genres']=movies['genres'].apply(collapse)
movies['keywords']=movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
movies['tags'][0]
```

```python
new=movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head()
```

```python
new['tags']=new['tags'].apply(lambda x:" ".join(x))
```

```python
new.head()
```

```python
new['tags']=new['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem("Heey my name")
```

```python
new['tags']=new['tags'].apply(stem)
```

```python
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new['tags']).toarray()
```

```python
vector[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similaraty=cosine_similarity(vector)
```

```python
similaraty
```

```python
new[new['title']=='Batman']
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similaraty[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python

recommend('Batman')
```
